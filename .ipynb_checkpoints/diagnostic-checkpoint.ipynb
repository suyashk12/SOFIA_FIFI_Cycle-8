```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from scipy import optimize
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import ScalarFormatter
```

```python
df = pd.read_csv("/Users/thepoetoftwilight/Desktop/emission_line_database.csv")
```

```python
df.dtypes
```

```python
df["o3_52"] = pd.to_numeric(df["o3_52"] , errors = "coerce")
df["err_52"] = pd.to_numeric(df["err_52"] , errors = "coerce")
df["n3_57"] = pd.to_numeric(df["n3_57"] , errors = "coerce")
df["err_57"] = pd.to_numeric(df["err_57"] , errors = "coerce")
```

```python
df.dtypes
```

```python
df
```

```python
Z = df["Z (Zsun)"]
o3_52 = df["o3_52"]
n3_57 = df["n3_57"]
o3_88 = df["o3_88"]
names = df["Galaxy"]
```

```python
logZ_Z_0 = np.log10(Z)
diagnostic = np.log10((2.2*o3_88 + o3_52)/n3_57)
```

```python
fig, ax = plt.subplots(figsize = (18, 13))

ax.scatter(logZ_Z_0[0:11], diagnostic[0:11], s = 100, color = 'red', label = "Herschel/PACS")
ax.scatter(logZ_Z_0[11:28], diagnostic[11:28], s = 100, color = 'green', label = "SOFIA/FIFI-LS+Herschel/PACS")
ax.scatter(logZ_Z_0[27:], diagnostic[27:], s = 100, color = 'blue', label = "SOFIA/FIFI-LS")

ax.set_xlabel(r'$\log(Z/Z_\odot)$', fontsize = 38, labelpad = 5)
ax.set_ylabel(r'$\log \left( \frac{2.2[OIII]_{58} + [OIII]_{52}}{[NIII]_{57}} \right)$', fontsize = 38, labelpad = 10)
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax.ticklabel_format(axis='y', style='sci', scilimits = (0, 0))
ax.tick_params(labelsize = 38, pad = 10)
ax.yaxis.offsetText.set_fontsize(36)

for i, txt in enumerate(names):
    ax.annotate(txt, (logZ_Z_0[i], diagnostic[i]), size = 15)
    
ax.legend(prop={'size': 20}, loc = 'upper right')
```
